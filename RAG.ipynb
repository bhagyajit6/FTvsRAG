```python
file_path = "medquad_cleaned.csv"
loader = CSVLoader(file_path=file_path)
data = loader.load()

for record in data[0]:
    print(record)
```

```python
len(data)
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=20)
docs = text_splitter.split_documents(data)
print(len(docs))
```

```python
embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

```python
from unsloth import FastLanguageModel
```

```python
models = ["unsloth/Qwen2.5-7B", 
          "unsloth/Meta-Llama-3.1-8B", 
          "unsloth/Phi-3.5-mini-instruct",
          "unsloth/gemma-2-9b",
          "unsloth/mistral-7b-instruct-v0.3-bnb-4bit"
          ]

fine_tuned_models = ["bpingua/Qwen2.5-7B-Medquad-16bit",
                     "bpingua/Llama-3.1-8B-medquad",
                     "bpingua/mistral-7b-instruct-v0.3-bnb-4bit-Medquad",
                     "bpingua/gemma-2-9b-medquad",
                     "bpingua/Phi-3.5-mini-instruct-Medquad"
                     ]

model_name = "unsloth/Qwen2.5-7B" 
model,tokenizer = FastLanguageModel.from_pretrained(
    model_name,
    load_in_4bit=True,  
    dtype=None
)
FastLanguageModel.for_inference(model)
```

```python
import torch
pipe = pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 256, 
                do_sample=True,
                top_k=30,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
```

```python
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0.5})
```
